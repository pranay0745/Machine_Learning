```python
# we need to fill NaNs before we can split.

df.rest_type.fillna('Unknown', inplace= True)
```

```python
rest_type_list= []
for rest_type in df.rest_type:
    
    rest_type_list.append(rest_type.split(',')[0])


df['rest_type_extract']= rest_type_list
```

```python
rest_type_count= df[df.rate != 0].groupby('rest_type_extract')['rest_type_extract'].count().sort_values(ascending= False)
rest_type_avg_rating= df[df.rate != 0].groupby('rest_type_extract')['rate'].mean().sort_values(ascending= False)
rest_type_max_rating= df[df.rate != 0].groupby('rest_type_extract')['rate'].max().sort_values(ascending= False)
rest_type_min_rating= df[df.rate != 0].groupby('rest_type_extract')['rate'].min().sort_values(ascending= False)
```

```python
print(df.dish_liked.isna().sum()) # we will fill the NaNs with "" (blanks)
df.dish_liked.fillna("", inplace= True)
print(df.dish_liked.isna().sum())
```

```python
liked_list= []

for dishes in df.dish_liked:
    
    dish= dishes.split(',')
    
    for words in dish:
        
        words= words.strip()
        
        liked_list.append(words)
```

```python
from collections import Counter
```

```python
top_dishes= Counter(liked_list)

top_10_dishes= top_dishes.most_common(11)[1:] #we will avoid the blank
```

```python
dishes= []
dish_counts= []

for items in top_10_dishes:
    
    dish, counts= items
    
    dishes.append(dish)
    dish_counts.append(counts)
```

```python
plt.figure(figsize= (24, 6))

sns.barplot(x= dishes, y= dish_counts)
plt.title('Top 10 liked dish as listed by Restaurants', fontsize= 'xx-large')
plt.xlabel('Dishes', fontsize= 'x-large')
plt.ylabel('Count', fontsize= 'x-large')
plt.xticks(fontsize= 'large')
```
